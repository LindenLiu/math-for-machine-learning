```python
x = np.random.randn(20)
y = np.random.randn(20)
z = np.random.randn(20)

# 3D drawing of the points
fig = plt.figure()
# add a 3D subplot
ax = fig.add_subplot(111, projection='3d')
# draw the points
ax.scatter(x, y, z)
# show the the figure
plt.show()
```

```python
from sklearn.datasets import load_digits


digits = load_digits()
data = digits.data
labels = digits.target
```

```python
idx = np.argsort(np.random.random(len(data)))

test_set_size = 200
x_test = data[idx[:test_set_size]]
y_test = labels[idx[:test_set_size]]
x_train = data[idx[test_set_size:]]
y_train = labels[idx[test_set_size:]]
```

```python
from sklearn.neural_network import MLPClassifier 

clf = MLPClassifier(hidden_layer_sizes=(128,))
clf.fit(x_train, y_train)
```

```python
from math import ceil


score = clf.score(x_test, y_test)
y_pred = clf.predict(x_test)
err = np.where(y_pred != y_test)[0]
print("Score: ", score)
print("Errors: ")
print("   actural: ", y_test[err])
print("   predicted: ", y_pred[err])
row = ceil(len(err) / 3) 
fig, axes = plt.subplots(int(row), 3, figsize=(8, 8))
for i, ax in enumerate(axes.flat):
    ax.imshow(x_test[err[i]].reshape(8, 8), cmap=plt.cm.gray_r)
    ax.set_title("Prediction: %i actual %i" % (y_pred[err[i]], y_test[err[i]]))
    if i == len(err) - 1:
        break
```
